# Chapter 17 - Getting Started with Vertical Slice Architecture

- Instead of seperating an application into layers, vertical slice encapsulates all horizontal concerns withing a single feature scope.
- Minimize coupling between slices
- Goal is for slices to be completly independent and isolated
    * Refactor reusable code as needed but no complex abrasctions up front

> The key point here is to code your handlers independently, organize them the best you can, keep an 
eye open for shared logic and emerging patterns, extract and encapsulate that logic, and try to limit 
interactions between use cases and slices.
